Import packages:

```python
import numpy as np
```

Load input:

```python
with open('input_day2.txt') as f:
    policies = []
    for line in f: # read lines
        policies.append([x for x in line.split()])
```

```python
policies
```

### Part 1
Check valid passwords:

```python
# Counter to check amount of valid passwords
valid_counter1 = 0

# Hardcoded:
#  - All lines have 3 elements: range, character, password
#  - All lines have a fixed range (x - y)
#  - All lines have 1 key alphabetic character

# Loop over policies
for p in policies:
    # split and finde range, cast t integer
    range_str = p[0].split('-')
    low = int(range_str[0])
    high = int(range_str[1])
    
    # count character in password (remove :)
    if p[2].count(p[1][0]) >= low and p[2].count(p[1][0]) <= high:
        valid_counter1 += 1
```

```python
valid_counter1
```

### Part 2
Check valid passwords

```python
# Counter to check amount of valid passwords
valid_counter2 = 0

# Hardcoded:
#  - All lines have 3 elements: range, character, password
#  - All lines have a fixed range (x - y)
#  - All lines have 1 key alphabetic character

# Loop over policies
for p in policies:
    # split and finde range, cast t integer
    range_str = p[0].split('-') # split
    first = int(range_str[0])-1     # cast to integer, subtract 1 for indexing start at 0
    second = int(range_str[1])-1
    
    print(p[1][0])
    
    # catch index out of range error and continue
    try:
        # check if first AND NOT second index contains element
        if p[2][first] == p[1][0]:
            if p[2][second] != p[1][0]: # split if statements to cast error after first check
                valid_counter2 += 1
        # check if NOT first and second index contains element
        elif p[2][first] != p[1][0]:
            if p[2][first] == p[1][0]:
                valid_counter2 += 1
    except IndexError:
        continue
```

```python
valid_counter2
```
