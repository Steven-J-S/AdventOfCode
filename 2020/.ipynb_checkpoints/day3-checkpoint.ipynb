### Part 1
Count trees

```python
count_trees(forest, down=1, step=3):
    # Check length of each row
    length = len(forest[0])-1 # max index range of element in row
    # init tree counter
    tree_counter = 0
    # init element indexer
    tree_index = 0

    for i in range(len(forest)):
        if tree_index >= length:
            tree_index -= length + 1
        if forest[i][tree_index] == '#':
            tree_counter += 1
        tree_index += 3
```
